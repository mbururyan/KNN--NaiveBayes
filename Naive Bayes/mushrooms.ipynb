```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklearn.naive_bayes import MultinomialNB, GaussianNB, BernoulliNB
from sklearn import metrics
```

```python
df = pd.read_csv('http://bit.ly/MushroomDataset')
df
```

```python
df.columns
```

```python
# Encode the dataframe

# Will change the columns to categorical datatype then encode them

def encoder(dataframe):
    for i in dataframe.columns:
        dataframe[i] = dataframe[i].astype('category')
        dataframe[i] = dataframe[i].cat.codes
    return dataframe.head()
        
```

```python
encoder(df)
```

```python
# Label encoder
from sklearn.preprocessing import LabelEncoder

encode = LabelEncoder()

for i in df.columns:
    df[i] = encode.fit_transform(df[i])
```

```python
df
```

```python
df
```

```python
df['cap-shape'].value_counts()
```

```python
X = df.iloc[:, -1].values.reshape(-1, 1)
y = df['mushroom'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Run it in a Naive Bayes Model

# Will use multinomial

multiNB = BernoulliNB().fit(X_train, y_train)
```

```python
y_pred = multiNB.predict(X_test)
```

```python
# Test for performance

print(metrics.accuracy_score(y_test, y_pred))
```

```python
print(metrics.classification_report(y_test, y_pred))
```
